```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../My_Datas/Data/meli.csv')
```

```python
df = pd.DataFrame(df)
df
```

```python
df['date_created'] = pd.to_datetime(df['date_created'], dayfirst=True)
```

```python
df['month'] = df['date_created'].dt.month
```

```python
valor_bruto = df.groupby('month')['transaction_amount'].sum()
```

```python
saldo_liquido = df.groupby('month')['net_received_amount'].sum()
saldo_liquido.reset_index()
```

```python
taxa = df.groupby('month')['TaxaML/MP'].sum()
taxa.reset_index()
```

```python
df_value = pd.DataFrame(valor_bruto)
```

```python
df_combined = pd.merge(df_value, saldo_liquido, on='month')
df_combined = pd.merge(df_combined, taxa, on='month')
```
