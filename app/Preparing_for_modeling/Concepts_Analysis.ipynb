```python
# Read all four files into a Jupyter Notebook
import pandas as pd

april = pd.read_excel('../My_Datas/Data/Book_Sales_April.xlsx')
may = pd.read_excel('../My_Datas/Data/Book_Sales_May.xlsx')
june = pd.read_excel('../My_Datas/Data/Book_Sales_June.xlsx')
customers = pd.read_csv('../My_Datas/Data/Book_Customers.csv')
```

```python
may.head()
```

```python
june.head()
```

```python
sales = pd.concat([april, may, june]).reset_index(drop=True)
```

```python
customers.head()
```

```python
sales.head()
```

```python
df = sales.merge(customers, how='left' ,on='Customer ID')
```

# Preparing Rows For Modeling


```python
df.head()
```

```python
june.head()
```

```python
june_purchases = (june.groupby('Customer ID')['Book']
 .count()
  .rename("June Purchases")
   .to_frame()
    .reset_index())

june_purchases.head()
```

```python
df_april_may = df[df["Purchase Date"].dt.month < 6]
df_april_may.tail()
```

```python
total_spend = (df_april_may.groupby('Customer ID')['Price']
 .sum()
 .rename('Total Spend')
 .to_frame()
 .reset_index())
total_spend.head()
```

```python
total_spend.merge(june_purchases, how="left").fillna(0)
```
