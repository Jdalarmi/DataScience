```python
import pandas as pd
```

```python
df = pd.read_json('../Explore_Data/Data/dados.json')
```

```python
df = pd.DataFrame(df)
```

```python
df_normalize = pd.json_normalize(df['data'])
```

```python
df_normalize
```

```python
Total_taxa = int(df_normalize["TaxaML/MP"].sum())
total_venda = int(df_normalize["transaction_amount"].sum())
total_recebido = int(df_normalize["net_received_amount"].sum())
```

```python
import matplotlib.pyplot as plt
```

```python
labels = ['Total', 'Recebido', 'Taxa MELI']
valores = [total_venda, total_recebido, Total_taxa]
```

```python
plt.bar(labels, valores, color=['blue', 'green', 'red'])
for i, v in enumerate(valores):
    plt.text(i, v + 50, str(v), ha='center', va='bottom')
    
plt.xlabel('')
plt.ylabel('Valor')
plt.title('Vendas MELI')

# Exibindo o gráfico
plt.show()
```
