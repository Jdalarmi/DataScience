```python
import pandas as pd 
import sqlite3

conn = sqlite3.connect('../My_Datas/Data/online_shop.db')
```

```python
#Views the transaction table
transactions = pd.read_sql("SELECT * FROM transactions", conn)
transactions.head()
```

```python
#Views the items table
items = pd.read_sql("SELECT * FROM items", conn)
items.head()
```

```python
#merge two tables together
df = transactions.merge(items, how="left", on='item_id')
df
```
